```python
# @title PATSTAT Explorer
import subprocess, os, sys
from pathlib import Path
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

REPO_URL = "https://github.com/mtcberlin/epo-tip-svelte-patstat-explorer.git"
APP_DIR = Path.home() / "patstat_svelte"

def _status(msg):
    clear_output(wait=True)
    display(HTML(f"""
    <div style="font-family: system-ui, sans-serif; padding: 20px;
                background: #f8fafc; border-left: 4px solid #4a5bc7;
                border-radius: 8px;">
        <div style="font-size: 15px; color: #334155;">⏳ {msg}</div>
    </div>"""))

def _launch(_=None):
    if (APP_DIR / ".git").is_dir():
        _status("Updating PATSTAT Explorer...")
        subprocess.run(["git", "-C", str(APP_DIR), "pull", "--ff-only"],
                       capture_output=True, text=True)
    else:
        _status("Downloading PATSTAT Explorer (first run)...")
        subprocess.run(["git", "clone", REPO_URL, str(APP_DIR)],
                       capture_output=True, text=True, check=True)
    os.chdir(str(APP_DIR))
    exec(open("launch.py").read(), globals())

# --- UI ---
clear_output()
display(HTML("""
<div style="font-family: system-ui, -apple-system, sans-serif; text-align: center;
            padding: 40px 20px; max-width: 520px; margin: 0 auto;">
    <div style="margin-bottom: 8px;">
        <span style="font-size: 32px; font-weight: 700; color: #2d3470;">PATSTAT Explorer</span>
    </div>
    <div style="font-size: 14px; color: #64748b; margin-bottom: 28px;">
        Patent analysis &amp; visualization on EPO PATSTAT
        <br/>
        <a href="https://mtc.berlin" target="_blank"
           style="color: #4a5bc7; text-decoration: none; font-weight: 600;">mtc.berlin</a>
        &middot; EPO TIP
    </div>
</div>
"""))

btn = widgets.Button(
    description='Start PATSTAT Explorer',
    button_style='',
    layout=widgets.Layout(width='280px', height='44px', margin='0 auto'),
    style={'button_color': '#4a5bc7', 'font_weight': 'bold'},
)
btn.add_class('patstat-btn')
btn.on_click(_launch)

display(HTML("""
<style>
    .patstat-btn button {
        color: white !important;
        font-size: 15px !important;
        border-radius: 8px !important;
        border: none !important;
        cursor: pointer !important;
        transition: opacity 0.2s !important;
    }
    .patstat-btn button:hover { opacity: 0.9 !important; }
</style>
"""))
display(btn)

display(HTML("""
<div style="font-family: system-ui, sans-serif; text-align: center;
            font-size: 12px; color: #94a3b8; margin-top: 16px;">
    First run downloads &amp; installs everything (~1–2 min).<br/>
    Subsequent starts are fast (~10s).<br/>
    To stop: run <code>stop()</code> or restart the kernel.
</div>
"""))
```
